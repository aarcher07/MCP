```python
import matplotlib.pyplot as plt
%matplotlib nbagg

import numpy as np
from matplotlib.widgets import Slider, Button, RadioButtons
Km = 1
Vmax = 1
C = np.linspace(0,50,100000)
Linear = np.divide(Vmax*C,Km)
Rate = np.divide(Vmax*C,Km+C)
fig, ax = plt.subplots(figsize = (9,6))
plt.subplots_adjust(bottom=0.25, right=0.8)

percentage = 10

full, = plt.plot(C,Rate, color = 'royalblue', lw = 2, label = '\nFull Rate\nEquations\n')
linear, = plt.plot(C,Linear, color = 'palevioletred', lw = 2, label = '\nLinear Rate\nEquations\n')
halfmax = plt.axvline(x=Km, color = "black", lw = 2, ls = '--', label = '\nHalf Max\nConcentration\n')
plt.xlim(0,1.5)
plt.ylim(0,1)
t = plt.axvline(x=0.01*percentage/Km, color = "seagreen", lw = 2, ls = '--', label = '\nConcentration\nat Error\nThreshold')


plt.xlabel('Unitless Concentration', fontsize = 15)
plt.ylabel('Unitless Reaction Rate', fontsize = 15)
axthresh = plt.axes([0.18, 0.1, 0.6, 0.03])
sthresh = Slider(axthresh, 'Error in\nLinear Approx', 0, 100.0, valinit=percentage, color = 'seagreen')
sthresh.vline.set_color('chartreuse')
sthresh.vline.set_linewidth(2)

cvalue = plt.text(70,10,"Threshold\nConcentration:\n\n{}".format(0.01*percentage/Km), fontsize = 15)

def update(val):
    test = sthresh.val
    t.set_xdata(x=(0.01*test/Km))
    concent = np.round(0.01*test/Km,3)
    cvalue.set_text("Threshold\nConcentration:\n\n{0}".format(concent))
    fig.canvas.draw_idle()
    
sthresh.on_changed(update)

resetax = plt.axes([0.6, 0.01, 0.1, 0.07])
button = Button(resetax, 'Reset', color='lightyellow', hovercolor='0.975')

def reset(event):
    sthresh.reset()
button.on_clicked(reset)

plt.legend(handles = [full, linear, halfmax, t], loc = [2.2,5])
plt.ylim([np.min(Rate), np.max(Rate)])
plt.xlim([0,1.5])
plt.show()
```
